```python
# Import recursive feature elimination
from sklearn.feature_selection import RFE

# Create the RFE object using two features and rank the features for the diabetes datasets
regr = LinearRegression()
rfe = RFE(estimator=regr, n_features_to_select=2, step=1)
rfe.fit(diabetes_X_train,diabetes_y_train)

#Calculate the mean squared error with the RFE object
print("Mean squared error with 2 features: %.2f"
      % np.mean((rfe.predict(diabetes_X_test) - diabetes_y_test) ** 2))
```

```python
# Imports
from sklearn.datasets import make_classification, load_iris
from sklearn import svm

# Load the dataset
iris = load_iris()

#Split the dataset
iris_X_train = iris.data[:-20]
iris_X_test = iris.data[-20:]

iris_y_train = iris.target[:-20]
iris_y_test = iris.target[-20:]
```

```python
# Make a linear SVM
clf_linear = svm.SVC(kernel='linear')
clf_linear.fit(iris_X_train, iris_y_train)

# Predict the class
pred = clf_linear.predict(iris_X_test)

print "Linear SVM predictions for iris predictions"
print "---------------------------------------------------------"

errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
        
print "  Test data errors = ",errors, "out of", len(pred), "predictions"


total_pred = clf_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != iris.target[i]:
        errors += 1.0

print "  Total errors = ",errors, "out of", len(total_pred), "predictions"
```

```python
# Create non-linear SVM
clf_non_linear = svm.SVC(kernel='rbf')
clf_non_linear.fit(iris_X_train,iris_y_train)
pred = clf_non_linear.predict(iris_X_test)

print "Non-linear SVM predictions for iris predictions"
print "---------------------------------------------------------"

errors = 0.0
for i in range(len(pred)):
    if pred[i] != iris_y_test[i]:
        errors += 1.0
print "   Test data errors = ",errors, "out of", len(pred), "predictions"

total_pred = clf_non_linear.predict(iris.data)
errors = 0.0
for i in range(len(total_pred)):
    if total_pred[i] != iris.target[i]:
        errors += 1.0
        
print "   Total errors = ",errors, "out of", len(total_pred), "predictions"
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(random_state=0) # Set random_state=0 for consistent prediction scores on each run
tree.fit(iris_X_train, iris_y_train)

print "DecisionTreeClassifier ---------------------------------------------"
print "   Prediction score for test data: ", tree.score(iris_X_test, iris_y_test)
print "   Prediction score for all data", tree.score(iris.data, iris.target)
```

```python
# Import kmeans algorithm and clustering evaluation metrics
from sklearn.cluster import KMeans
from sklearn import metrics

# Load the dataset
iris = load_iris()

# Use K-means on it
kmeans = KMeans(init='random', n_clusters = 3, n_init=10)
kmeans.fit(iris.data)
    
# Homogeneity score
print "Homogeneity_score = ", metrics.homogeneity_score(iris.target,kmeans.labels_)
    
# Completeness score
print "Completeness_score = ", metrics.completeness_score(iris.target,kmeans.labels_)
```
